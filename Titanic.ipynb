```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# EDA 

```python
data=pd.read_csv("train_and_test2.csv")
data.head(5)
```

```python
data.describe()
```

```python
data=data[["Passengerid","Age","Fare","Sex","sibsp","Pclass","Embarked","2urvived"]]
data.head()
```

```python
data=data.drop(["Passengerid"],axis=1)
data.head()
```

```python
data=data.rename(columns={"2urvived":"Survived"})
data.head()
```

```python
data.isnull().sum()
```

```python
data=data.dropna()
```

```python
data.isnull().sum()
```

# Analyze

```python
sns.countplot(x="Pclass",data=data,color="purple")
```

```python
sns.countplot(x="Sex",data=data,color="orange")
```

```python
sns.countplot(x="Survived",data=data)
```

```python
grouped_data = data.groupby('Pclass')['Fare'].mean().reset_index()
sns.barplot(x='Pclass', y='Fare', data=grouped_data, color='skyblue')
```

```python
survival_rate=data.groupby("Pclass")["Survived"].mean().reset_index()
labels = ["First Class" if pclass == 1 else "Second Class" if pclass == 2 else "Third Class" for pclass in survival_rate["Pclass"]]
sizes=survival_rate["Survived"]*100
```

```python
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['skyblue', 'yellow', 'lightcoral'], startangle=140)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
saved=data.groupby("Sex")["Survived"].mean().reset_index()
labels = ["Men" if sex == 0 else "Women" for sex in saved["Sex"]]
sizes=saved["Survived"]*100
```

```python
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['skyblue', 'purple'], startangle=140)
plt.title('Survival Rate by Gender')
plt.show()
```

# Logistic Regression 

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
x=data.drop(["Survived"],axis=1)
y=data["Survived"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
lr=LogisticRegression(max_iter=1000)
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,roc_curve, roc_auc_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
acc=accuracy_score(y_test,y_pred)
print(acc)
```

```python
fpr,tpr,_=roc_curve(y_test,y_pred)
roc_auc=roc_auc_score(y_test,y_pred)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(fpr, tpr, color='blue', label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=40)
rfc.fit(x_train,y_train)
```

```python
y_pred=rfc.predict(x_test)
acc=accuracy_score(y_test,y_pred)
print(acc)
```
